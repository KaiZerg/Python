```python
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup as BS
```

```python
r = requests.get("https://web.archive.org/web/20211220133941/https://zniis.ru/statistics/")
html = BS(r.content, 'html.parser')
```

```python
data = []
```

```python
for td in html('td'):
    if td.text:
        data.append(td.text)
    else:
        print("else")
```

```python
print(data)
```

Находим индекс первого дня года(01.01.2021) в нашем списке, которое является последней строкой нашей будущей таблицы для анализа

```python
print(data.index('01.01.2021'))
```

находим первое упоминание "IV квартал"

```python
print(data.index('IV квартал'))
```

```python
# разделяю список по индексу последней цифры будущей таблицы
relevant_data = data[:1059]
print(relevant_data)
```

Разделяю список на 3 списка, будущие колонки

```python
dates = relevant_data[0::3]
dates
```

```python
daily_new_requests = relevant_data[1::3]
daily_new_requests
```

```python
daily_transferred_numbers = relevant_data[2::3]
daily_transferred_numbers
```

Функция для удаления пробелов в строках

```python
def remove(string):
    return "".join(string.split())
```

Функция перевода строки в дату(не работает)

```python
from datetime import datetime
def format_date(date):
    y = datetime.strptime(date, '%d.%m.%Y').strftime("%d.%m.%Y")
    return y
```

```python
from datetime import datetime
project_date = datetime.strptime(dates[1], '%d.%m.%Y').strftime("%d.%m.%Y")
print(project_date)
print(type(project_date))
```

```python
dates = [datetime.strptime(x, '%d.%m.%Y').strftime("%d.%m.%Y") for x in dates]
dates
```

```python
daily_new_requests = [int(remove(x)) for x in daily_new_requests]
daily_new_requests
```

```python
# расчет ежедневного прироста заявок
diff_list_requests = []
for x, y in zip(daily_new_requests[0::], daily_new_requests[1::]):
    diff_list_requests.append(x-y)
diff_list_requests
```

```python
daily_transferred_numbers = [int(remove(x)) for x in daily_transferred_numbers]
daily_transferred_numbers
```

```python
# расчет ежедневного прироста перенесенных номеров
diff_list_transferred_numbers = []
for x, y in zip(daily_transferred_numbers[0::], daily_transferred_numbers[1::]):
    diff_list_transferred_numbers.append(x-y)
diff_list_transferred_numbers
```

Добавляю в конец списка расчетное значение для 01.01.2021 - среднее между двумя последними элементами.

```python
# расчет ежедневного прироста заявок
diff_list_requests.append((diff_list_requests[-1] + diff_list_requests[-2]) / 2)
diff_list_requests
```

```python
diff_list_transferred_numbers.append((diff_list_transferred_numbers[-1] + diff_list_transferred_numbers[-2]) / 2)
diff_list_transferred_numbers
```

Проверяю длину всех списков, перед соединением в таблицу

```python
print(len(dates))
print(len(daily_new_requests))
print(len(diff_list_requests))
print(len(daily_transferred_numbers))
print(len(diff_list_transferred_numbers))
```

собираю все нужные списки в один DataFrame

```python
series_dates = pd.Series(dates)

series_requests = pd.Series(diff_list_requests)

series_transfers = pd.Series(diff_list_transferred_numbers)

df_dates= pd.DataFrame(series_dates)
df_dates = df_dates.rename(columns = {0:'дата'})

df_requests = pd.DataFrame(series_requests)
df_requests = df_requests.rename(columns = {0:'ежедневный прирост заявок'})

df_transfers = pd.DataFrame(series_transfers)
df_transfers = df_transfers.rename(columns = {0:'ежедневный прирост перенесенных'})

df_all = pd.concat([df_dates, df_requests, df_transfers ], axis=1)
print(df_all)
print(type(df_all))
```

Сохраняю таблицу в Excel файл

```python
writer = pd.ExcelWriter('2021.xlsx') 
df_all.to_excel(writer)
writer.save()
```
